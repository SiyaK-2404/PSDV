### Q3)Fibonacci Sequence

```python
import matplotlib.pyplot as plt
n=int(input())
def fibonacci(n):
    fibo_sequence = [0, 1]
    
    for i in range(2, n):
        next_fibo = fibo_sequence[-1] + fibo_sequence[-2]
        fibo_sequence.append(next_fibo)
    
    if n == 1:
        return [0]
    
    return fibo_sequence

print(fibonacci(n))
fibo = fibonacci(n)
plt.plot(range(1, n + 1), fibo)
plt.title('Fibonacci Sequence')
plt.xlabel('Index (n)')
plt.ylabel('Fibonacci Number')
plt.show()
```

### Q10)Character Frequency

```python
import matplotlib.pyplot as plt
def character_frequency(s):
    s = s.lower()
    
    frequency = {}
    
    for char in s:
        frequency[char] = frequency.get(char, 0) + 1
    
    return frequency
s=input()
print(character_frequency(s))
```

```python
frequency = character_frequency(s)
x = frequency.keys()
y = frequency.values()
plt.bar(x,y,color="purple",edgecolor="black")
plt.title("Bar Chart")
plt.xlabel("Character")
plt.ylabel("Frequency")
plt.show()
```

### Q11)Remove Duplicates

```python
def remove_duplicates(n):
    unique_numbers=[]
    for i in n:
        if i not in unique_numbers:
            unique_numbers.append(i)

    mean_of_unique_numbers=sum(unique_numbers)/len(unique_numbers)

    unique_numbers.sort()
    length = len(unique_numbers)
    if length % 2 == 0:
        median_of_unique_numbers = (unique_numbers[length // 2 - 1] + unique_numbers[length // 2]) / 2
    else:
        median_of_unique_numbers = unique_numbers[length // 2]

    return unique_numbers,  mean_of_unique_numbers,median_of_unique_numbers
t=input()
numbers = list(map(int, t.split()))
result = remove_duplicates(numbers)
print(result[0],result[1],result[2])
```

### Q12)String Operations

```python
import matplotlib.pyplot as plt
def analyze_strings(string_list):
    longest_string = ""
    vowel_count = 0
    vowels = {'a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U'}
    
    for s in string_list:
        if len(s) > len(longest_string):
            longest_string = s
        
        if s[0] in vowels:
            vowel_count += 1
    
    return longest_string, vowel_count
string_list=input()
string_list = [s.strip() for s in string_list.split(',')]
print(analyze_strings(string_list))

string_lengths = [len(s) for s in string_list]
plt.hist(string_lengths,bins=range(min(string_lengths), max(string_lengths) + 2), color="orange",edgecolor="black")
plt.title('Distribution of String Lengths')
plt.xlabel('Length of String')
plt.ylabel('Frequency')
plt.show()
    
```

### Q13)Sorting Dictionaries

```python
def sort_by_key(data, key):
    for i in range(len(data)):
        for j in range(0, len(data) - i - 1):
            if data[j][key] > data[j + 1][key]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

data=input()
data=eval(data)
sorted_data = sort_by_key(data, 'age')
print(sorted_data)
```

### Q18)Word Count and Frequency

```python
import matplotlib.pyplot as plt
def remove_punctuation(text):
    punctuation = ".,;!?()[]{}\"':"
    no_punctuation_text = ""
    for char in text:
        if char not in punctuation:
            no_punctuation_text += char
    return no_punctuation_text

with open("sample.txt", 'r') as file:
    content = file.read()


final_content = remove_punctuation(content)


words = final_content.split()

word_count = {}
for word in words:
    word = word.lower() 
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

most_frequent_word = None
max_count = 0

for word, count in word_count.items():
    if count > max_count:
        max_count = count
        most_frequent_word = word

num_words = len(words)
print(f"Number of words: {num_words}")
print(f"Most frequent word: '{most_frequent_word}' with frequency: {max_count}")
```

```python
x = list(word_count.keys()) 
y = list(word_count.values())
plt.figure(figsize=(18, 8))
plt.bar(x,y,color="lime",edgecolor="black")
plt.title("Bar Graph")
plt.xlabel("Words")
plt.ylabel("Frequency of Words")
plt.tight_layout()

plt.show()
```

### Q22)Normalization

```python
import matplotlib.pyplot as plt
def normalization(number):
    min_num=min(number)
    max_num=max(number)
    return [(x - min_num) / (max_num - min_num) for x in number]

number = [10, 20, 30, 40, 50]
normalized_values=normalization(number)
print(normalized_values)
```

```python
#Line plot for visualisation of normalised data
plt.plot(number,normalized_values,color="red")
plt.title("Line Plot")
plt.xlabel("Numbers")
plt.ylabel("Normalized Values")
plt.show()
```

### Q25)Leap Year Checker

```python
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

year = int(input())
print(is_leap_year(year))
```
